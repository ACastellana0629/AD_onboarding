## Step 3: Clean DXmerge diagnosis/outcome variables
clean DXmerge_data and create the first version of our outcome labels.

```r
dx_clean <- DXmerge_data %>%
  mutate(
    scan_date = mdy(scandate),

    # Four-group outcome:
    # AD = already Dementia at baseline
    # CONVERTER = CN/MCI at baseline, later Dementia
    # CN = CN baseline, no later Dementia
    # MCI = MCI baseline, no later Dementia
    group4 = case_when(
      DX.new == "Dement" ~ "AD",
      DX.new %in% c("CN", "MCI") & ad_status == "Dement" ~ "CONVERTER",
      DX.new == "CN" & ad_status == "No" ~ "CN",
      DX.new == "MCI" & ad_status == "No" ~ "MCI",
      TRUE ~ NA_character_
    ),

    # Binary converter variable for later secondary analysis
    converter_binary = case_when(
      DX.new %in% c("CN", "MCI") & ad_status == "Dement" ~ 1,
      DX.new %in% c("CN", "MCI") & ad_status == "No" ~ 0,
      TRUE ~ NA_real_
    )
  ) %>%
  select(
    PTID,
    scan_date,
    RID,
    SITEID,
    DX.new,
    ad_status,
    ad_time,
    group4,
    converter_binary,
    APOE4,
    E4_combine,
    E4_3grp,
    PTGENDER,
    Age_at_scan,
    PTEDUCAT,
    PTHAND,
    PTMARRY,
    PTETHCAT,
    PTRACCAT
  )

# Check the cleaned DXmerge data
dim(dx_clean)
head(dx_clean)

# Check group counts
table(dx_clean$DX.new, useNA = "ifany")
table(dx_clean$ad_status, useNA = "ifany")
table(dx_clean$group4, useNA = "ifany")
table(dx_clean$converter_binary, useNA = "ifany")
```

# Step 12: Quality Control (QC) for Cortical Thickness Values

Before fitting prediction models, I performed a light quality-control check on the cortical thickness variables.

## Why this QC step is needed
Cortical thickness values in the DKT atlas should generally fall in a biologically plausible range (roughly around 1–5 mm). While exploring the long-format cortical thickness data, I found a very small number of implausibly low values, including one value of 0 and two values below 1 mm. These are unlikely to reflect true anatomy and are more likely to be segmentation / preprocessing artifacts.

Because cortical thickness is the main predictor set in this analysis, even a few impossible values could distort:
- regional mean thickness estimates,
- group comparisons across CN / MCI / CONVERTER / AD,
- logistic regression coefficients, and
- variable selection in the lasso model.

## QC rule used here
For the **68 cortical thickness predictors only**:

- if **thickness < 1 mm**, set that value to **NA**

I chose **not** to drop entire subjects at this stage, because only a very small number of region-level values were affected. A light-touch rule preserves the rest of each subject’s cortical thickness profile while removing clearly implausible values.

## Important note
This QC step applies **only to cortical thickness variables**. It does **not** change the diagnosis labels, converter definition, APOE4, age, or other demographic variables.

```r
# ============================================================
# Step 12: Quality control for cortical thickness values
# ============================================================

# Summary of thickness values across all long-format rows
summary(analysis_long$thickness)

# Count suspicious values
sum(analysis_long$thickness <= 0, na.rm = TRUE)
sum(analysis_long$thickness < 1, na.rm = TRUE)
sum(analysis_long$thickness > 5, na.rm = TRUE)

# Show rows with impossible or suspicious thickness values
bad_thickness_rows <- analysis_long %>%
  filter(thickness <= 0 | thickness < 1 | thickness > 5) %>%
  arrange(thickness)

bad_thickness_rows
```

Used a light-touch QC rule:

Rule

For the 68 cortical thickness predictors only:

if thickness < 1, set it to NA

I would not drop entire subjects at this point, because:

only 2 subjects are affected
only 3 region-level values are suspicious
dropping a whole subject because of 1–2 bad regions is too aggressive for a first pass

```r
# ============================================================
# Step 13: Apply QC fix to cortical thickness values
# Rule: set thickness < 1 mm to NA
# ============================================================

analysis_long_qc <- analysis_long %>%
  mutate(
    thickness = ifelse(thickness < 1, NA, thickness)
  )

# Check that suspicious values are now removed
summary(analysis_long_qc$thickness)
sum(analysis_long_qc$thickness < 1, na.rm = TRUE)

analysis_long_qc %>%
  filter(is.na(thickness)) %>%
  select(PTID, scan_date, group4, hemisphere, region, region_full, thickness)
```

Then we need to rebuild the wide summary table from the QC’d data

```r
# ============================================================
# Step 13b: Recompute region summaries after QC
# ============================================================

region_summary_qc <- analysis_long_qc %>%
  group_by(group4, hemisphere, region, region_full) %>%
  summarise(
    n_nonmissing = sum(!is.na(thickness)),
    mean_thickness = mean(thickness, na.rm = TRUE),
    sd_thickness = sd(thickness, na.rm = TRUE),
    median_thickness = median(thickness, na.rm = TRUE),
    .groups = "drop"
  )

region_summary_wide_qc <- region_summary_qc %>%
  select(group4, hemisphere, region, region_full, mean_thickness) %>%
  pivot_wider(
    names_from = group4,
    values_from = mean_thickness
  ) %>%
  mutate(
    AD_minus_CN = AD - CN,
    CONVERTER_minus_CN = CONVERTER - CN,
    CONVERTER_minus_MCI = CONVERTER - MCI
  )

# Preview top AD-vs-CN thinning regions after QC
region_summary_wide_qc %>%
  arrange(AD_minus_CN) %>%
  head(15)
```

# Step 14: Region-wise ANOVA Screening

Before fitting a prediction model, we first test whether each cortical thickness region differs across the four clinical groups: CN, MCI, CONVERTER, and AD.

For each of the 68 cortical thickness predictors, we fit a simple one-way ANOVA model:

$thickness \sim group4$

This asks whether mean cortical thickness differs across the four groups for each brain region.

This step is useful because it gives us a ranked list of regions that show the strongest group-level differences. It also helps justify which regions may be useful for later prediction models.

Because we test 68 regions, we apply false discovery rate correction to control for multiple comparisons.

```r
# ============================================================
# Step 14: One-way ANOVA for each cortical thickness region
# Model: thickness ~ group4
# ============================================================

anova_results <- analysis_long %>%
  group_by(region_full, hemisphere, region) %>%
  group_modify(~{
    fit <- aov(thickness ~ group4, data = .x)
    fit_summary <- summary(fit)[[1]]

    tibble(
      df_group = fit_summary["group4", "Df"],
      df_resid = fit_summary["Residuals", "Df"],
      F_value = fit_summary["group4", "F value"],
      p_value = fit_summary["group4", "Pr(>F)"]
    )
  }) %>%
  ungroup() %>%
  mutate(
    p_adj_fdr = p.adjust(p_value, method = "fdr")
  ) %>%
  arrange(p_value)

# View top ANOVA results
anova_results %>%
  select(region_full, hemisphere, region, F_value, p_value, p_adj_fdr) %>%
  head(20)
```

```r
# ============================================================
# Step 14b: Summarize significant ANOVA results
# ============================================================

# Number of regions significant before FDR correction
sum(anova_results$p_value < 0.05, na.rm = TRUE)

# Number of regions significant after FDR correction
sum(anova_results$p_adj_fdr < 0.05, na.rm = TRUE)

# Top 15 regions by ANOVA p-value
top_anova_regions <- anova_results %>%
  select(region_full, hemisphere, region, F_value, p_value, p_adj_fdr) %>%
  arrange(p_value) %>%
  head(15)

top_anova_regions
```

# Step 15: First Converter Prediction Model

After identifying cortical regions that differ across the four clinical groups, we next fit a simple binary logistic regression model to predict conversion among participants who were CN or MCI at baseline.

The outcome variable is:

- **Converter** = participant later developed dementia/AD
- **Non-converter** = participant did not later develop dementia/AD

For the first model, we use a small set of biologically relevant cortical thickness predictors rather than all 68 regions. This keeps the model interpretable and reduces the risk of overfitting, since the number of converters is relatively small.

The first model includes:

- left entorhinal thickness
- right entorhinal thickness
- left temporal pole thickness
- right temporal pole thickness
- age at scan

These regions were selected because they showed some of the strongest group differences in the exploratory and ANOVA analyses and are also strongly implicated in Alzheimer’s disease.

```r
# ============================================================
# Step 15: Logistic regression model for converter prediction
# Outcome: converter_binary
# Predictors: bilateral entorhinal + bilateral temporal pole + age
# ============================================================

converter_model_data <- analysis_converter %>%
  select(
    converter_binary,
    Age_at_scan,
    lh_entorhinal_thickness,
    rh_entorhinal_thickness,
    lh_temporalpole_thickness,
    rh_temporalpole_thickness
  ) %>%
  drop_na()

dim(converter_model_data)
head(converter_model_data)

# Fit logistic regression
converter_glm <- glm(
  converter_binary ~ Age_at_scan +
    lh_entorhinal_thickness +
    rh_entorhinal_thickness +
    lh_temporalpole_thickness +
    rh_temporalpole_thickness,
  data = converter_model_data,
  family = binomial
)

summary(converter_glm)
```

```r
# ============================================================
# Step 15b: Convert coefficients to odds ratios
# ============================================================

converter_or_table <- broom::tidy(converter_glm) %>%
  mutate(
    odds_ratio = exp(estimate),
    conf.low.or = exp(estimate - 1.96 * std.error),
    conf.high.or = exp(estimate + 1.96 * std.error)
  )

converter_or_table
```

```r
# ============================================================
# Step 15c: Simple in-sample prediction summary
# ============================================================

converter_model_data <- converter_model_data %>%
  mutate(
    pred_prob = predict(converter_glm, type = "response"),
    pred_class = ifelse(pred_prob >= 0.5, "Converter", "Non-converter"),
    pred_class = factor(pred_class, levels = c("Non-converter", "Converter"))
  )

# Confusion matrix
table(
  Predicted = converter_model_data$pred_class,
  Actual = converter_model_data$converter_binary
)

# Mean predicted probability by true class
converter_model_data %>%
  group_by(converter_binary) %>%
  summarise(
    mean_pred_prob = mean(pred_prob),
    median_pred_prob = median(pred_prob),
    .groups = "drop"
  )
```

# Step 16: Simplified Converter Model Using Entorhinal Thickness

The first logistic regression suggested that entorhinal thickness carries the strongest signal for conversion risk, while temporal pole thickness contributed little once entorhinal measures were included.

To make the model more interpretable and reduce unnecessary complexity, we next fit a simplified logistic regression model using only:

- age at scan
- left entorhinal thickness
- right entorhinal thickness

This model is easier to interpret and focuses on the cortical region that showed the strongest evidence of group differences in the earlier exploratory and ANOVA analyses.

```r
# ============================================================
# Step 16: Simplified logistic regression model
# Outcome: converter_binary
# Predictors: age + bilateral entorhinal thickness
# ============================================================

converter_glm_ent <- glm(
  converter_binary ~ Age_at_scan +
    lh_entorhinal_thickness +
    rh_entorhinal_thickness,
  data = converter_model_data,
  family = binomial
)

summary(converter_glm_ent)
```

```r
# ============================================================
# Step 16b: Odds ratios for simplified model
# ============================================================

converter_glm_ent_or <- broom::tidy(converter_glm_ent) %>%
  mutate(
    odds_ratio = exp(estimate),
    conf.low.or = exp(estimate - 1.96 * std.error),
    conf.high.or = exp(estimate + 1.96 * std.error)
  )

converter_glm_ent_or
```

```r
# ============================================================
# Step 16c: Compare full and simplified converter models
# ============================================================

AIC(converter_glm, converter_glm_ent)

anova(converter_glm_ent, converter_glm, test = "Chisq")
```

# Step 17: Evaluate the Simplified Converter Model

Because the converter outcome is relatively rare in this dataset, a default classification threshold of 0.5 is not ideal for evaluating model performance.

Instead, we next examine:

This provides a better sense of whether entorhinal thickness meaningfully separates converters from non-converters, even if few subjects cross a 0.5 probability threshold.

```r
# ============================================================
# Step 17b: Plot predicted probabilities by true converter status
# ============================================================

ggplot(converter_model_data, aes(x = converter_binary, y = pred_prob_ent)) +
  geom_boxplot(outlier.alpha = 0.4) +
  labs(
    title = "Predicted Conversion Probability by True Converter Status",
    x = "Observed Converter Status",
    y = "Predicted Probability of Conversion"
  ) +
  theme_minimal()
```

# Step 21: Higher-Dimensional Converter Prediction Using All 68 Cortical Thickness Features

The earlier converter models focused on a small, interpretable set of predictors (entorhinal thickness and APOE4). However, the broader project goal is to treat conversion as a prediction task using the cortical thickness features from both hemispheres.

To take a first step in that direction, I fit a penalized logistic regression (lasso) model using all 68 cortical thickness predictors to classify **converter vs non-converter** among baseline CN/MCI participants.

I use lasso regularization because the number of cortical predictors is large relative to the number of converters, and lasso provides both shrinkage and automatic variable selection.

```r
# ============================================================
# Step 21: Lasso logistic regression using all 68 cortical thickness features
# Outcome: converter vs non-converter
# Predictors: all 68 cortical thickness variables
# ============================================================

library(glmnet)
library(dplyr)
library(tidyr)
library(pROC)

# ------------------------------------------------------------
# 1) Define cortical thickness columns directly from analysis_converter
# ------------------------------------------------------------
thickness_cols <- names(analysis_converter)[grepl("^(lh|rh)_.*_thickness$", names(analysis_converter))]

length(thickness_cols)
head(thickness_cols)

# sanity check: should be 68
print(length(thickness_cols))

# ------------------------------------------------------------
# 2) Build lasso dataset
# Keep only rows with known converter status + complete thickness data
# ------------------------------------------------------------
lasso_data <- analysis_converter %>%
  filter(!is.na(converter_binary)) %>%
  select(converter_binary, all_of(thickness_cols)) %>%
  drop_na()

dim(lasso_data)
table(lasso_data$converter_binary)

# ------------------------------------------------------------
# 3) Create outcome vector and predictor matrix
# ------------------------------------------------------------
y_lasso <- ifelse(lasso_data$converter_binary == "Converter", 1, 0)
x_lasso <- as.matrix(lasso_data %>% select(-converter_binary))

dim(x_lasso)
length(y_lasso)
```

```r
# ============================================================
# Step 21b: Cross-validated lasso
# ============================================================

set.seed(123)

cv_lasso <- cv.glmnet(
  x = x_lasso,
  y = y_lasso,
  family = "binomial",
  alpha = 1,           # lasso penalty
  nfolds = 10,
  type.measure = "auc"
)

plot(cv_lasso)

cv_lasso$lambda.min
cv_lasso$lambda.1se
max(cv_lasso$cvm, na.rm = TRUE)
```

```r
# ============================================================
# Step 21c: Variables selected by lasso
# Using lambda.1se = more conservative / easier to explain
# ============================================================

lasso_coef <- coef(cv_lasso, s = "lambda.1se")

lasso_coef_df <- as.matrix(lasso_coef) %>%
  as.data.frame() %>%
  tibble::rownames_to_column("term")

names(lasso_coef_df)[2] <- "coef"

lasso_coef_df <- lasso_coef_df %>%
  filter(coef != 0)

lasso_coef_df
```

```r
# ============================================================
# Step 21d: Lasso predicted probabilities and AUC
# ============================================================

lasso_pred <- predict(
  cv_lasso,
  newx = x_lasso,
  s = "lambda.1se",
  type = "response"
)[,1]

roc_lasso <- roc(
  response = lasso_data$converter_binary,
  predictor = lasso_pred,
  levels = c("Non-converter", "Converter"),
  direction = "<"
)

auc(roc_lasso)
```

The lasso model gives two AUC values:

- The cross-validated AUC from `cv.glmnet`, which is the more honest performance estimate.
- The in-sample AUC from predicting on the same data used to fit the model, which is optimistic and should not be treated as final performance.

For reporting, I should emphasize the cross-validated AUC.

### Interpreting the lasso model performance

The AUC reported above is the **in-sample AUC**, meaning it is calculated on the same data used to fit the lasso model. This can be optimistic and should not be treated as the final estimate of out-of-sample predictive performance.

For discussion in the meeting, the more important quantity is the **cross-validated AUC** returned by `cv.glmnet`, since that better reflects how the model may generalize to new data.

So I view the lasso analysis as a **first high-dimensional prototype** using all 68 cortical thickness predictors, rather than a finalized predictive model.

```r
# ============================================================
# Step 21e: Lasso model performance summary
# ============================================================

lasso_cv_auc <- max(cv_lasso$cvm, na.rm = TRUE)
lasso_insample_auc <- as.numeric(auc(roc_lasso))

lasso_summary <- tibble(
  model = "All 68 cortical thickness predictors: lasso logistic regression",
  n = nrow(lasso_data),
  lambda_min = cv_lasso$lambda.min,
  lambda_1se = cv_lasso$lambda.1se,
  cv_auc = lasso_cv_auc,
  insample_auc = lasso_insample_auc
)

lasso_summary
```

## Step 24: Preparing the Four-Class Classification Dataset

Before fitting any predictive models, the dataset was organized into four clinically meaningful outcome groups representing different stages of cognitive decline:

The resulting datasets contain:

  * MCI: 160

```r
# ============================================================
# Step 24a: Check four-class outcome and cortical predictors
# ============================================================

library(dplyr)
library(tidyr)

# Confirm outcome counts
analysis_4group %>%
  count(group4)

# Confirm cortical thickness predictors
length(brain_thickness_cols)

brain_thickness_cols
```

## Step 25c — Cross-validated Multinomial LASSO

Instead of choosing the penalty (λ) ourselves, we'll let the data choose it using 10-fold cross-validation.

This is standard practice and makes the baseline much more defensible.

```r
# ============================================================
# Step 25c: Cross-validated Multinomial LASSO
# ============================================================

set.seed(123)

cv_multinom <- cv.glmnet(
  x = X_train_matrix,
  y = y_train,
  family = "multinomial",
  alpha = 1,          # LASSO
  nfolds = 10,
  type.measure = "class"
)

# Plot cross-validation curve
plot(cv_multinom)
```

These are extremely important.

This usually gives the highest predictive performance.

λ_1se

The largest lambda whose error is within one standard error of the minimum.

This gives:

fewer variables
simpler model
often nearly identical accuracy

Many published biomedical papers report results using λ_1se because it produces a more parsimonious model.

## Step 25: Baseline Four-Class Prediction Model

The objective of this step was to establish a baseline machine learning model capable of predicting one of four clinical groups using only regional cortical thickness measurements derived from the DKT atlas.

### Outcome Classes

The response variable consisted of four diagnostic categories:

- **CN:** Cognitively Normal
- **MCI:** Stable Mild Cognitive Impairment
- **CONVERTER:** Participants who were cognitively normal or had MCI at baseline but later progressed to Alzheimer's disease (dementia)
- **AD:** Participants diagnosed with Alzheimer's disease (dementia) at baseline

The predictor variables consisted exclusively of the **68 regional cortical thickness measurements**, with participant identifiers and scan dates excluded from model training to avoid information leakage.

### Model Selection

A **penalized multinomial logistic regression (LASSO)** was chosen as the baseline classifier.

Unlike ordinary multinomial logistic regression, LASSO introduces a regularization penalty that shrinks less informative coefficients toward zero. This is particularly useful because many cortical thickness measurements are highly correlated across neighboring brain regions. Regularization helps reduce overfitting while simultaneously identifying the regions that contribute most to disease classification.

### Model Training

The model was trained using the **80% training dataset (606 participants)** while the remaining **20% (154 participants)** was held out for independent evaluation.

Rather than manually selecting the regularization parameter (λ), a **10-fold cross-validation** procedure was used. During cross-validation, the training data were repeatedly partitioned into smaller training and validation folds, and prediction error was computed across a range of λ values. The λ producing the lowest cross-validation error (`lambda.min`) and the simpler λ within one standard error of the minimum (`lambda.1se`) were both identified.

For this analysis, **`lambda.1se`** was selected because it produces a more parsimonious model while maintaining predictive performance.

### Selected Cortical Regions

After regularization, only a small subset of the original 68 cortical thickness measurements retained non-zero coefficients. These selected regions represent the cortical areas that contributed most strongly to distinguishing the four diagnostic groups.

Several selected regions—including the **entorhinal cortex, fusiform gyrus, inferior parietal cortex, and superior temporal cortex**—were also identified during the earlier exploratory analyses as regions exhibiting substantial cortical thinning. The agreement between the exploratory analysis and the predictive model suggests that these cortical regions may serve as informative biomarkers of Alzheimer's disease progression.

The next step is to evaluate the trained model on the held-out test set using a confusion matrix and classification accuracy to determine how well cortical thickness alone can distinguish among the four clinical groups.

# Step 26: Model Evaluation on the Held-Out Test Set

After training the penalized multinomial logistic regression model using the training dataset, the next step is to evaluate how well the model generalizes to previously unseen participants.

The model is applied to the independent **20% testing dataset**, which was not used during model training or cross-validation. This allows the predictive performance to be evaluated without introducing optimistic bias.

One of the primary goals of this analysis is to investigate whether regional cortical thickness measurements alone are sufficient to distinguish among the four clinical groups.
