## Step 3: Clean DXmerge diagnosis/outcome variables
clean DXmerge_data and create the first version of our outcome labels.

```r
dx_clean <- DXmerge_data %>%
  mutate(
    scan_date = mdy(scandate),

    # Four-group outcome:
    # AD = already Dementia at baseline
    # CONVERTER = CN/MCI at baseline, later Dementia
    # CN = CN baseline, no later Dementia
    # MCI = MCI baseline, no later Dementia
    group4 = case_when(
      DX.new == "Dement" ~ "AD",
      DX.new %in% c("CN", "MCI") & ad_status == "Dement" ~ "CONVERTER",
      DX.new == "CN" & ad_status == "No" ~ "CN",
      DX.new == "MCI" & ad_status == "No" ~ "MCI",
      TRUE ~ NA_character_
    ),

    # Binary converter variable for later secondary analysis
    converter_binary = case_when(
      DX.new %in% c("CN", "MCI") & ad_status == "Dement" ~ 1,
      DX.new %in% c("CN", "MCI") & ad_status == "No" ~ 0,
      TRUE ~ NA_real_
    )
  ) %>%
  select(
    PTID,
    scan_date,
    RID,
    SITEID,
    DX.new,
    ad_status,
    ad_time,
    group4,
    converter_binary,
    APOE4,
    E4_combine,
    E4_3grp,
    PTGENDER,
    Age_at_scan,
    PTEDUCAT,
    PTHAND,
    PTMARRY,
    PTETHCAT,
    PTRACCAT
  )

# Check the cleaned DXmerge data
dim(dx_clean)
head(dx_clean)

# Check group counts
table(dx_clean$DX.new, useNA = "ifany")
table(dx_clean$ad_status, useNA = "ifany")
table(dx_clean$group4, useNA = "ifany")
table(dx_clean$converter_binary, useNA = "ifany")
```

# Step 14: Region-wise ANOVA Screening

Before fitting a prediction model, we first test whether each cortical thickness region differs across the four clinical groups: CN, MCI, CONVERTER, and AD.

For each of the 68 cortical thickness predictors, we fit a simple one-way ANOVA model:

$thickness \sim group4$

This asks whether mean cortical thickness differs across the four groups for each brain region.

This step is useful because it gives us a ranked list of regions that show the strongest group-level differences. It also helps justify which regions may be useful for later prediction models.

Because we test 68 regions, we apply false discovery rate correction to control for multiple comparisons.

```r
# ============================================================
# Step 14: One-way ANOVA for each cortical thickness region
# Model: thickness ~ group4
# ============================================================

anova_results <- analysis_long %>%
  group_by(region_full, hemisphere, region) %>%
  group_modify(~{
    fit <- aov(thickness ~ group4, data = .x)
    fit_summary <- summary(fit)[[1]]

    tibble(
      df_group = fit_summary["group4", "Df"],
      df_resid = fit_summary["Residuals", "Df"],
      F_value = fit_summary["group4", "F value"],
      p_value = fit_summary["group4", "Pr(>F)"]
    )
  }) %>%
  ungroup() %>%
  mutate(
    p_adj_fdr = p.adjust(p_value, method = "fdr")
  ) %>%
  arrange(p_value)

# View top ANOVA results
anova_results %>%
  select(region_full, hemisphere, region, F_value, p_value, p_adj_fdr) %>%
  head(20)
```

```r
# ============================================================
# Step 14b: Summarize significant ANOVA results
# ============================================================

# Number of regions significant before FDR correction
sum(anova_results$p_value < 0.05, na.rm = TRUE)

# Number of regions significant after FDR correction
sum(anova_results$p_adj_fdr < 0.05, na.rm = TRUE)

# Top 15 regions by ANOVA p-value
top_anova_regions <- anova_results %>%
  select(region_full, hemisphere, region, F_value, p_value, p_adj_fdr) %>%
  arrange(p_value) %>%
  head(15)

top_anova_regions
```

# Step 15: First Converter Prediction Model

After identifying cortical regions that differ across the four clinical groups, we next fit a simple binary logistic regression model to predict conversion among participants who were CN or MCI at baseline.

The outcome variable is:

- **Converter** = participant later developed dementia/AD
- **Non-converter** = participant did not later develop dementia/AD

For the first model, we use a small set of biologically relevant cortical thickness predictors rather than all 68 regions. This keeps the model interpretable and reduces the risk of overfitting, since the number of converters is relatively small.

The first model includes:

- left entorhinal thickness
- right entorhinal thickness
- left temporal pole thickness
- right temporal pole thickness
- age at scan

These regions were selected because they showed some of the strongest group differences in the exploratory and ANOVA analyses and are also strongly implicated in Alzheimer’s disease.

```r
# ============================================================
# Step 15: Logistic regression model for converter prediction
# Outcome: converter_binary
# Predictors: bilateral entorhinal + bilateral temporal pole + age
# ============================================================

converter_model_data <- analysis_converter %>%
  select(
    converter_binary,
    Age_at_scan,
    lh_entorhinal_thickness,
    rh_entorhinal_thickness,
    lh_temporalpole_thickness,
    rh_temporalpole_thickness
  ) %>%
  drop_na()

dim(converter_model_data)
head(converter_model_data)

# Fit logistic regression
converter_glm <- glm(
  converter_binary ~ Age_at_scan +
    lh_entorhinal_thickness +
    rh_entorhinal_thickness +
    lh_temporalpole_thickness +
    rh_temporalpole_thickness,
  data = converter_model_data,
  family = binomial
)

summary(converter_glm)
```

```r
# ============================================================
# Step 15b: Convert coefficients to odds ratios
# ============================================================

converter_or_table <- broom::tidy(converter_glm) %>%
  mutate(
    odds_ratio = exp(estimate),
    conf.low.or = exp(estimate - 1.96 * std.error),
    conf.high.or = exp(estimate + 1.96 * std.error)
  )

converter_or_table
```

```r
# ============================================================
# Step 15c: Simple in-sample prediction summary
# ============================================================

converter_model_data <- converter_model_data %>%
  mutate(
    pred_prob = predict(converter_glm, type = "response"),
    pred_class = ifelse(pred_prob >= 0.5, "Converter", "Non-converter"),
    pred_class = factor(pred_class, levels = c("Non-converter", "Converter"))
  )

# Confusion matrix
table(
  Predicted = converter_model_data$pred_class,
  Actual = converter_model_data$converter_binary
)

# Mean predicted probability by true class
converter_model_data %>%
  group_by(converter_binary) %>%
  summarise(
    mean_pred_prob = mean(pred_prob),
    median_pred_prob = median(pred_prob),
    .groups = "drop"
  )
```

# Step 16: Simplified Converter Model Using Entorhinal Thickness

The first logistic regression suggested that entorhinal thickness carries the strongest signal for conversion risk, while temporal pole thickness contributed little once entorhinal measures were included.

To make the model more interpretable and reduce unnecessary complexity, we next fit a simplified logistic regression model using only:

- age at scan
- left entorhinal thickness
- right entorhinal thickness

This model is easier to interpret and focuses on the cortical region that showed the strongest evidence of group differences in the earlier exploratory and ANOVA analyses.

```r
# ============================================================
# Step 16: Simplified logistic regression model
# Outcome: converter_binary
# Predictors: age + bilateral entorhinal thickness
# ============================================================

converter_glm_ent <- glm(
  converter_binary ~ Age_at_scan +
    lh_entorhinal_thickness +
    rh_entorhinal_thickness,
  data = converter_model_data,
  family = binomial
)

summary(converter_glm_ent)
```

```r
# ============================================================
# Step 16b: Odds ratios for simplified model
# ============================================================

converter_glm_ent_or <- broom::tidy(converter_glm_ent) %>%
  mutate(
    odds_ratio = exp(estimate),
    conf.low.or = exp(estimate - 1.96 * std.error),
    conf.high.or = exp(estimate + 1.96 * std.error)
  )

converter_glm_ent_or
```

```r
# ============================================================
# Step 16c: Compare full and simplified converter models
# ============================================================

AIC(converter_glm, converter_glm_ent)

anova(converter_glm_ent, converter_glm, test = "Chisq")
```

# Step 17: Evaluate the Simplified Converter Model

Because the converter outcome is relatively rare in this dataset, a default classification threshold of 0.5 is not ideal for evaluating model performance.

Instead, we next examine:

1. the predicted probabilities from the simplified logistic regression model,
2. how those probabilities differ between converters and non-converters, and
3. the ROC curve and AUC as threshold-independent measures of discrimination.

This provides a better sense of whether entorhinal thickness meaningfully separates converters from non-converters, even if few subjects cross a 0.5 probability threshold.

```r
# ============================================================
# Step 17: Predicted probabilities from simplified model
# ============================================================

converter_model_data <- converter_model_data %>%
  mutate(
    pred_prob_ent = predict(converter_glm_ent, type = "response")
  )

# Summary of predicted probabilities by true class
converter_model_data %>%
  group_by(converter_binary) %>%
  summarise(
    n = n(),
    mean_pred_prob = mean(pred_prob_ent),
    median_pred_prob = median(pred_prob_ent),
    sd_pred_prob = sd(pred_prob_ent),
    min_pred_prob = min(pred_prob_ent),
    max_pred_prob = max(pred_prob_ent),
    .groups = "drop"
  )
```

```r
# ============================================================
# Step 17b: Plot predicted probabilities by true converter status
# ============================================================

ggplot(converter_model_data, aes(x = converter_binary, y = pred_prob_ent)) +
  geom_boxplot(outlier.alpha = 0.4) +
  labs(
    title = "Predicted Conversion Probability by True Converter Status",
    x = "Observed Converter Status",
    y = "Predicted Probability of Conversion"
  ) +
  theme_minimal()
```

```r
library(pROC)
```

```r
# ============================================================
# Step 17c: ROC curve and AUC for simplified converter model
# ============================================================

roc_ent <- pROC::roc(
  response = converter_model_data$converter_binary,
  predictor = converter_model_data$pred_prob_ent,
  levels = c("Non-converter", "Converter"),
  direction = "<"
)

auc(roc_ent)

plot(
  roc_ent,
  main = "ROC Curve: Simplified Entorhinal Converter Model"
)
```

```r
# ============================================================
# Step 17d: Classification at a lower threshold
# ============================================================

threshold <- 0.10

converter_model_data <- converter_model_data %>%
  mutate(
    pred_class_010 = ifelse(pred_prob_ent >= threshold, "Converter", "Non-converter"),
    pred_class_010 = factor(pred_class_010, levels = c("Non-converter", "Converter"))
  )

table(
  Predicted = converter_model_data$pred_class_010,
  Actual = converter_model_data$converter_binary
)
```
