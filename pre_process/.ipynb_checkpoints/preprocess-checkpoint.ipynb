### Review missingness data in APOE4

- Already noticed APOE4 has missing values

```python
missing = (
    data.isnull()
        .sum()
        .sort_values(ascending=False)
)

missing_pct = (
    100 * data.isnull().mean()
).sort_values(ascending=False)

pd.DataFrame({
    "Missing Count": missing,
    "Missing Percent": missing_pct
})
```

## Missingness Summary

The largest missingness appears in `ad_status` and `ad_time`, with 263 missing values each, corresponding to about 28.8% of the dataset. This suggests that AD conversion/status outcome information may only be available for a subset of participants.

The APOE-related variables (`APOE4`, `E4_3grp`, and `E4_combine`) are each missing 121 values, or about 13.3% of the dataset. Because APOE4 is biologically important in Alzheimer's disease risk, these missing values should be tracked carefully before modeling.

Diagnosis-related fields (`DX.new`, `DX.EXAMDATE`, `VISCODE2`, `daydiff`, and `mindaydiff`) are missing for only 7 participants, less than 1% of the dataset.

```python
data[["DX.new", "APOE4", "ad_status", "ad_time"]].isnull().sum()
```

** key observation is that ad_status and ad_time are missing for exactly the same 263 participants, while APOE4 has a different missingness pattern. **

```python
pd.crosstab(
    data["DX.new"],
    data["APOE4"].isna()
)
```

### APOE4 Missingness by Diagnosis

APOE4 missingness appears relatively similar across diagnostic groups:

- CN: 54 / 481 (~11.2%)
- MCI: 49 / 315 (~15.6%)
- Dementia: 11 / 110 (~10.0%)

No strong evidence was observed that APOE4 missingness is concentrated in a particular diagnostic category. This suggests that APOE4 missingness may be approximately independent of baseline diagnosis, although formal statistical testing would be required to confirm this.

### APOE4 Prevalence by Diagnosis

The prevalence of APOE4 positivity increases across diagnostic categories:

| Diagnosis | APOE4 Positive |
|------------|------------|
| CN | 31.6% |
| MCI | 42.9% |
| Dementia | 66.7% |

This pattern is consistent with the established role of APOE ε4 as a major genetic risk factor for Alzheimer's disease. APOE4-positive individuals are substantially more common in the dementia group than in cognitively normal participants.

The observed trend suggests that the diagnostic labels and genetic information are biologically consistent and provides an important sanity check for subsequent analyses.

```python
import pandas as pd
import matplotlib.pyplot as plt

apoe_rates = (
    pd.crosstab(
        data["DX.new"],
        data["APOE4"],
        normalize="index"
    )["P"] * 100
)

apoe_rates.plot(kind="bar")

plt.ylabel("APOE4 Positive (%)")
plt.xlabel("Diagnosis")
plt.title("APOE4 Positivity by Diagnosis")
plt.show()
```

### Age Distribution by Diagnosis

Mean age increases slightly across diagnostic groups:

| Diagnosis | Mean Age (years) |
|------------|------------|
| CN | 73.21 |
| MCI | 73.78 |
| Dementia | 75.95 |

Participants with dementia are, on average, approximately 2.7 years older than cognitively normal participants. While this trend is consistent with the increased prevalence of Alzheimer's disease at older ages, the overlap between age distributions remains substantial.

This suggests that age may act as a confounding variable in downstream analyses and should likely be included as a covariate when modeling disease-related effects.

### Age Distribution Across Diagnostic Groups

The boxplot demonstrates substantial overlap in age distributions across CN, MCI, and Dementia participants.

Although participants with dementia tend to be older on average, age alone does not clearly separate diagnostic groups. The median age increases gradually from CN to MCI to Dementia, suggesting a modest association between age and disease status.

Because age differs across groups while still exhibiting substantial overlap, age should likely be considered a potential confounding variable in downstream analyses.

### Preliminary Interpretation of Longitudinal Outcomes

The variable `ad_status` does not appear to represent baseline diagnosis, as its values consist primarily of `No` and `Dement`. Instead, it likely represents a longitudinal outcome indicating whether a participant eventually converted to dementia during follow-up.

The accompanying variable `ad_time` is available for exactly the same subset of participants (650 individuals), suggesting that it may encode time-to-conversion or time-to-last-follow-up.

These observations suggest that the dataset may have been designed for predicting future dementia progression from baseline demographic, genetic, and imaging measurements.

### Relationship Between Baseline Diagnosis and Longitudinal Outcome

Cross-tabulation between baseline diagnosis (`DX.new`) and longitudinal outcome (`ad_status`) suggests that `ad_status` represents future dementia conversion rather than baseline disease status.

Key observations:

- Only 4 cognitively normal (CN) participants converted to dementia.
- 53 MCI participants converted to dementia.
- Participants already diagnosed with dementia at baseline have missing values for `ad_status`, which is consistent with the outcome being undefined for individuals who already have dementia.

These findings strongly suggest that:

- `DX.new` represents baseline diagnosis.
- `ad_status` represents future dementia conversion.
- `ad_time` likely represents time-to-conversion or time-to-last-follow-up.

This indicates that the dataset may have been designed for dementia progression prediction using baseline demographic, genetic, imaging, and clinical variables.

## MCI Conversion Analysis

Based on the lab meeting notes, one key goal is to focus on participants diagnosed with MCI at baseline and distinguish between:

- Stable MCI: baseline MCI participants who did not convert to dementia during follow-up.
- MCI converters: baseline MCI participants who later converted to dementia.

This setup reframes the problem from baseline diagnosis classification to longitudinal prediction: can baseline information predict future dementia conversion among MCI participants?

```python
mci_data = data[data["DX.new"] == "MCI"].copy()

mci_data["conversion_status"] = mci_data["ad_status"].map({
    "No": "Stable MCI",
    "Dement": "MCI Converter"
})

mci_data["conversion_status"].value_counts(dropna=False)
```

```python
pd.crosstab(
    mci_data["conversion_status"],
    mci_data["APOE4"],
    dropna=False
)
```

```python
mci_data.groupby("conversion_status")["Age_at_scan"].describe()
```

### Working Prediction Cohort

The likely prediction cohort is the subset of participants with baseline MCI and known longitudinal outcomes.

Participants with missing `ad_status` should be excluded from the first supervised prediction analysis, because their conversion status is unknown.

The preliminary target variable is:

- 0 = Stable MCI
- 1 = MCI converter

Candidate predictors include age, sex, education, APOE4 status, and eventually rs-fMRI features.

### Baseline Characteristics of MCI Converters

Among participants with baseline MCI and known follow-up outcomes:

#### APOE4

- Stable MCI: ~39.8% APOE4 positive
- MCI Converters: ~56.9% APOE4 positive

APOE4 positivity is substantially higher among converters, consistent with APOE ε4 being a major genetic risk factor for Alzheimer's disease progression.

#### Age

- Stable MCI mean age: 73.1 years
- MCI Converter mean age: 75.3 years

Converters are slightly older on average, suggesting age may contribute to conversion risk.

These observations indicate that baseline demographic and genetic variables already contain predictive information regarding future dementia conversion.
