```python
## Open the data and ensure it was properly imported
import pandas as pd


data = pd.read_csv("../Data/dxmerge_rsfmri_baselinedata_2022_08_22.csv")

print(data.shape)
data.head()
```

```python
import os
print(os.getcwd())
```

```python
# Diagnosis distribution
# How many cognitively normal (CN), How many MCI, How many Dementia
data["DX.new"].value_counts(dropna=False)
```

```python
#visualization

data["DX.new"].value_counts().plot(kind="bar")
plt.title("Baseline Diagnosis Distribution")
plt.ylabel("Count")
plt.show()
```

### Age distribution by diagnosis

Let's see whether dementia subjects are older.

```python
data.groupby("DX.new")["Age_at_scan"].describe()
#visualization
data.boxplot(column="Age_at_scan", by="DX.new")
plt.title("Age at Scan by Diagnosis")
plt.suptitle("")
plt.ylabel("Age")
plt.show()
```

```python
#APOE4 and Diagnosis
data["APOE4"].value_counts(dropna=False)
pd.crosstab(
    data["DX.new"],
    data["APOE4"],
    normalize="index"
)
```
